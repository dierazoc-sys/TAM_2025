```python
# Calcula el producto punto entre vector_a y vector_b
dot_product = np.dot(vector_a, vector_b)

# Calcula el cuadrado de la norma de vector_b
norm_b_squared = np.dot(vector_b, vector_b)

# Calcula el escalar de proyección
projection_scalar = dot_product / norm_b_squared

# Multiplica el escalar por vector_b para obtener el vector proyectado analíticamente
projected_vector_analytical = projection_scalar * vector_b

# Muestra el vector proyectado
print(f"El vector proyectado analíticamente es: {projected_vector_analytical}")
```

### Subtask:

Implement the gradient descent algorithm to find the optimal scalar for the projection by defining the cost function and its derivative, initializing parameters, and iterating to update the scalar.



```python
# Define la función de costo (error al cuadrado entre el vector original y su proyección)
def cost_function(k, vector_a, vector_b):
  return np.sum((vector_a - k * vector_b)**2)

# Define la derivada de la función de costo con respecto a k
def cost_function_derivative(k, vector_a, vector_b):
  return -2 * np.sum((vector_a - k * vector_b) * vector_b)

# Inicializa un valor inicial para el escalar k
k = 0.0

# Define un learning rate (tasa de aprendizaje) pequeño
learning_rate = 0.01

# Define el número de iteraciones para el gradiente descendente
n_iterations = 1000

# Implementa un bucle que se ejecute durante el número especificado de iteraciones
for i in range(n_iterations):
  # Calcula el gradiente de la función de costo en el valor actual de k
  gradient = cost_function_derivative(k, vector_a, vector_b)

  # Actualiza k restando el producto del learning rate y el gradiente del valor actual de k
  k = k - learning_rate * gradient

# Después del bucle, multiplica el escalar k optimizado por vector_b para obtener el vector proyectado iterativamente
projected_vector_iterative = k * vector_b

# Imprime el vector proyectado iterativamente
print(f"El vector proyectado iterativamente es: {projected_vector_iterative}")
```

```python
# Este script demuestra la proyección de un vector en 2 dimensiones sobre otro
# utilizando tanto la solución analítica como la solución iterativa (gradiente descendente).

import numpy as np
import matplotlib.pyplot as plt

# ==============================================================================
# 1. Definición de los vectores
# ==============================================================================

# Define el primer vector en 2D.
# Este es el vector que queremos proyectar.
vector_a = np.array([3, 4])

# Define el segundo vector en 2D.
# Este es el vector sobre el cual proyectaremos vector_a.
vector_b = np.array([1, 0])

# ==============================================================================
# 2. Solución analítica
# ==============================================================================

# La proyección analítica de un vector 'a' sobre un vector 'b' se calcula usando la fórmula:
# proj_b(a) = ((a . b) / ||b||^2) * b
# donde '.' es el producto punto y '||b||^2' es la norma al cuadrado del vector b.

# Calcula el producto punto entre vector_a y vector_b.
# El producto punto (dot product) es una operación fundamental en álgebra lineal.
dot_product = np.dot(vector_a, vector_b)

# Calcula el cuadrado de la norma (magnitud) de vector_b.
# La norma al cuadrado es el producto punto del vector consigo mismo.
norm_b_squared = np.dot(vector_b, vector_b)

# Calcula el escalar de proyección.
# Este escalar representa cuánto "escalamos" el vector_b para obtener la proyección.
projection_scalar_analytical = dot_product / norm_b_squared

# Multiplica el escalar por vector_b para obtener el vector proyectado analíticamente.
# Este es el vector resultante de la proyección.
projected_vector_analytical = projection_scalar_analytical * vector_b

# Muestra el vector proyectado analíticamente.
print(f"El vector proyectado analíticamente es: {projected_vector_analytical}")

# ==============================================================================
# 3. Solución iterativa (Gradiente Descendente)
# ==============================================================================

# Utilizamos gradiente descendente para encontrar el escalar 'k' que minimiza el error
# entre el vector_a y la proyección k * vector_b.
# La función de costo es el error cuadrático: Cost(k) = ||vector_a - k * vector_b||^2

# Define la función de costo (error al cuadrado entre el vector original y su proyección).
# Esta función mide qué tan "cerca" está la proyección actual (k * vector_b) del vector original (vector_a).
def cost_function(k, vector_a, vector_b):
  return np.sum((vector_a - k * vector_b)**2)

# Define la derivada de la función de costo con respecto a k.
# La derivada nos indica la dirección de máximo incremento de la función de costo.
# Para minimizar la función de costo, nos movemos en la dirección opuesta al gradiente.
# La derivada de ||a - kb||^2 con respecto a k es -2 * (a - kb) . b
def cost_function_derivative(k, vector_a, vector_b):
  return -2 * np.sum((vector_a - k * vector_b) * vector_b)

# Inicializa un valor inicial para el escalar k.
# Comenzamos con una estimación inicial para el escalar de proyección.
k = 0.0

# Define un learning rate (tasa de aprendizaje) pequeño.
# El learning rate determina el tamaño del paso que damos en la dirección opuesta al gradiente.
# Un learning rate pequeño puede llevar a una convergencia más lenta pero más estable.
learning_rate = 0.01

# Define el número de iteraciones para el gradiente descendente.
# Este es el número de veces que actualizaremos el valor de k.
n_iterations = 1000

# Implementa un bucle que se ejecute durante el número especificado de iteraciones.
# En cada iteración, calculamos el gradiente y actualizamos k.
for i in range(n_iterations):
  # Calcula el gradiente de la función de costo en el valor actual de k.
  gradient = cost_function_derivative(k, vector_a, vector_b)

  # Actualiza k restando el producto del learning rate y el gradiente del valor actual de k.
  # Esta es la regla de actualización del gradiente descendente.
  k = k - learning_rate * gradient

# Después del bucle, multiplica el escalar k optimizado por vector_b para obtener el vector proyectado iterativamente.
# Este es el vector proyectado utilizando el escalar encontrado por gradiente descendente.
projected_vector_iterative = k * vector_b

# Imprime el vector proyectado iterativamente.
print(f"El vector proyectado iterativamente es: {projected_vector_iterative}")

# ==============================================================================
# 4. Visualización
# ==============================================================================

# Crea una figura y ejes para la gráfica.
fig, ax = plt.subplots(figsize=(8, 6))

# Graficar cada vector usando quiver.
# La función quiver dibuja flechas representando los vectores desde un punto de origen.
ax.quiver(0, 0, vector_a[0], vector_a[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector A')
ax.quiver(0, 0, vector_b[0], vector_b[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector B')
# Grafica el vector proyectado obtenido analíticamente.
ax.quiver(0, 0, projected_vector_analytical[0], projected_vector_analytical[1], angles='xy', scale_units='xy', scale=1, color='g', label='Vector Proyectado (Analítico)')
# Opcionalmente, puedes graficar el vector proyectado obtenido iterativamente para comparar.
# ax.quiver(0, 0, projected_vector_iterative[0], projected_vector_iterative[1], angles='xy', scale_units='xy', scale=1, color='purple', label='Vector Proyectado (Iterativo)')


# Establecer los límites de los ejes.
# Aseguramos que todos los vectores sean visibles en la gráfica.
max_limit = max(np.max(np.abs(vector_a)), np.max(np.abs(vector_b)), np.max(np.abs(projected_vector_analytical))) + 1
ax.set_xlim([-max_limit, max_limit])
ax.set_ylim([-max_limit, max_limit])

# Añadir etiquetas y título a la gráfica.
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_title('Proyección de Vector en 2D')

# Añadir leyenda para identificar los vectores.
ax.legend()

# Activar cuadrícula para facilitar la lectura de la gráfica.
ax.grid()

# Mostrar la gráfica.
plt.show()
```

## Solución iterativa (gradiente descendente)

### Subtask:
Define la función de costo (error al cuadrado entre el vector original y su proyección). Implementa el algoritmo de gradiente descendente para minimizar la función de costo y encontrar el escalar óptimo para la proyección.

**Reasoning**:
Implement the gradient descent algorithm to find the optimal scalar for the projection by defining the cost function and its derivative, initializing parameters, and iterating to update the scalar.

```python
# Define la función de costo (error al cuadrado entre el vector original y su proyección)
def cost_function(k, vector_a, vector_b):
  # La función de costo mide la diferencia cuadrática entre el vector original (vector_a)
  # y su proyección aproximada (k * vector_b). Queremos minimizar esta diferencia.
  return np.sum((vector_a - k * vector_b)**2)

# Define la derivada de la función de costo con respecto a k
def cost_function_derivative(k, vector_a, vector_b):
  # La derivada nos indica la dirección y magnitud del cambio en la función de costo
  # con respecto a 'k'. La fórmula se deriva de la regla de la cadena.
  return -2 * np.sum((vector_a - k * vector_b) * vector_b)

# Inicializa un valor inicial para el escalar k
# Empezamos con una estimación arbitraria para el escalar de proyección.
k = 0.0

# Define un learning rate (tasa de aprendizaje) pequeño
# El learning rate controla el tamaño de los pasos que da el algoritmo
# en la dirección opuesta al gradiente. Un valor pequeño ayuda a evitar
# sobrepasar el mínimo.
learning_rate = 0.01

# Define el número de iteraciones para el gradiente descendente
# Este es el número de veces que el algoritmo actualizará el valor de 'k'
# para acercarse al mínimo de la función de costo.
n_iterations = 1000

# Lista para almacenar los valores de la función de costo en cada iteración
cost_history = []

# Implementa un bucle que se ejecute durante el número especificado de iteraciones
for i in range(n_iterations):
  # Calcula el gradiente de la función de costo en el valor actual de k
  # Esto nos dice la pendiente de la función de costo en el punto actual 'k'.
  gradient = cost_function_derivative(k, vector_a, vector_b)

  # Actualiza k restando el producto del learning rate y el gradiente del valor actual de k
  # Este es el paso clave del gradiente descendente: movemos 'k' en la dirección
  # opuesta al gradiente para disminuir la función de costo.
  k = k - learning_rate * gradient

  # Almacena el valor de la función de costo en la iteración actual
  cost_history.append(cost_function(k, vector_a, vector_b))


# Después del bucle, multiplica el escalar k optimizado por vector_b para obtener el vector proyectado iterativamente
# Una vez que el algoritmo ha convergido (o completado las iteraciones), usamos el valor final de 'k'
# para calcular el vector proyectado.
projected_vector_iterative = k * vector_b

# Imprime el vector proyectado iterativamente
print(f"El vector proyectado iterativamente es: {projected_vector_iterative}")
```

### Subtask:
Grafica la función de costo en cada iteración del gradiente descendente para visualizar la convergencia.

**Reasoning**:
Plot the cost history collected during the gradient descent iterations to visualize the convergence of the algorithm.

```python
# Crea una figura y ejes para la gráfica de costo
fig_cost, ax_cost = plt.subplots(figsize=(8, 6))

# Grafica el historial de costos
ax_cost.plot(range(n_iterations), cost_history)

# Añade etiquetas y título
ax_cost.set_xlabel('Número de Iteraciones')
ax_cost.set_ylabel('Función de Costo')
ax_cost.set_title('Convergencia de la Función de Costo en el Gradiente Descendente')

# Activa cuadrícula
ax_cost.grid(True)

# Muestra la gráfica
plt.show()
```

Resize the uploaded images to a consistent size, convert the resized images to grayscale, convert the grayscale images into vectors, and normalize the image vectors.

Process the uploaded images by resizing, converting to grayscale, vectorizing, and normalizing them as described in the instructions.



```python
from PIL import Image
import io

# Predefined consistent size for resizing
target_size = (100, 100)

# List to store the flattened image vectors
image_vectors = []

# Iterate through the uploaded image files
for filename, content in uploaded.items():
  # Open the image using Pillow
  img = Image.open(io.BytesIO(content))

  # Resize the image
  img_resized = img.resize(target_size)

  # Convert the resized image to grayscale
  img_gray = img_resized.convert('L')

  # Convert the grayscale image to a NumPy array and flatten
  img_vector = np.array(img_gray).flatten()

  # Append the flattened vector to the list
  image_vectors.append(img_vector)

# Convert the list of image vectors to a NumPy array
image_vectors_np = np.array(image_vectors)

# Normalize the image vectors
# Divide each element by the maximum pixel value (255 for grayscale)
normalized_image_vectors = image_vectors_np / 255.0

# Print the shape of the normalized vectors to verify
print(f"Shape of normalized image vectors: {normalized_image_vectors.shape}")
```

## Iterative euclidean distance

### Subtask:
Implement an iterative method (like gradient descent) to find the Euclidean distance between the normalized image vectors.


**Reasoning**:
Define the cost function and its derivative for gradient descent, select two normalized vectors, initialize an intermediate vector and parameters, and implement the gradient descent loop to find the iterative Euclidean distance. This covers steps 1 through 7 and part of step 8 of the instructions.



```python
# 1. Define the function that calculates the squared Euclidean distance
def squared_euclidean_distance(vec1, vec2):
  """Calculates the squared Euclidean distance between two vectors."""
  return np.sum((vec1 - vec2)**2)

# 2. Define the derivative of the cost function
def squared_euclidean_distance_derivative(intermediate_vec, target_vec):
  """Calculates the derivative of the squared Euclidean distance with respect to the intermediate vector."""
  return 2 * (intermediate_vec - target_vec)

# 3. Choose two normalized image vectors
vector_a_iterative = normalized_image_vectors[0]
vector_b_iterative = normalized_image_vectors[1]

# 4. Initialize an "intermediate" vector
# We want to find a vector 'intermediate_vec' such that the distance between 'intermediate_vec' and 'vector_b_iterative'
# is minimized, which should be equal to the distance between 'vector_a_iterative' and 'vector_b_iterative'.
# For finding the distance iteratively, we can think of it as finding the vector 'v' such that ||vector_a_iterative - vector_b_iterative|| = ||v||.
# However, the subtask asks to "find the Euclidean distance between the normalized image vectors iteratively".
# A common way to use gradient descent for distance is to minimize the squared difference between the distances.
# Let's re-interpret the subtask: we will use gradient descent to find the difference vector (vector_a_iterative - vector_b_iterative)
# and then calculate its norm.
# The cost function to minimize would be ||(vector_a_iterative - vector_b_iterative) - v||^2 where v is the vector we are iteratively finding.
# The derivative with respect to v would be -2 * ((vector_a_iterative - vector_b_iterative) - v).
# Minimizing this cost will make v approach (vector_a_iterative - vector_b_iterative).

# Let's redefine the iterative approach to find the difference vector directly.
# Cost function: C(v) = ||(vector_a_iterative - vector_b_iterative) - v||^2
# Derivative with respect to v: dC/dv = -2 * (vector_a_iterative - vector_b_iterative - v)

# Initialize the vector we are iteratively finding (the difference vector approximation)
difference_vector_iterative = np.zeros_like(vector_a_iterative)

# 5. Initialize a learning rate and a number of iterations
learning_rate_iterative = 0.001 # Adjusted learning rate for this problem
n_iterations_iterative = 5000

# 6. Implement a loop for the gradient descent
for i in range(n_iterations_iterative):
  # Calculate the gradient of the cost function with respect to the intermediate vector
  gradient_iterative = -2 * ((vector_a_iterative - vector_b_iterative) - difference_vector_iterative)

  # Update the intermediate vector
  difference_vector_iterative = difference_vector_iterative - learning_rate_iterative * gradient_iterative

# 7. After the iterations, the Euclidean distance is the norm of the found difference vector
iterative_distance_1_2 = np.linalg.norm(difference_vector_iterative)

print(f"Iterative Euclidean distance between Image 1 and Image 2: {iterative_distance_1_2}")

# Repeat for other pairs (part of step 8)
# Image 1 and Image 3
vector_a_iterative_1_3 = normalized_image_vectors[0]
vector_b_iterative_1_3 = normalized_image_vectors[2]
difference_vector_iterative_1_3 = np.zeros_like(vector_a_iterative_1_3)

for i in range(n_iterations_iterative):
  gradient_iterative_1_3 = -2 * ((vector_a_iterative_1_3 - vector_b_iterative_1_3) - difference_vector_iterative_1_3)
  difference_vector_iterative_1_3 = difference_vector_iterative_1_3 - learning_rate_iterative * gradient_iterative_1_3

iterative_distance_1_3 = np.linalg.norm(difference_vector_iterative_1_3)
print(f"Iterative Euclidean distance between Image 1 and Image 3: {iterative_distance_1_3}")

# Image 2 and Image 3
vector_a_iterative_2_3 = normalized_image_vectors[1]
vector_b_iterative_2_3 = normalized_image_vectors[2]
difference_vector_iterative_2_3 = np.zeros_like(vector_a_iterative_2_3)

for i in range(n_iterations_iterative):
  gradient_iterative_2_3 = -2 * ((vector_a_iterative_2_3 - vector_b_iterative_2_3) - difference_vector_iterative_2_3)
  difference_vector_iterative_2_3 = difference_vector_iterative_2_3 - learning_rate_iterative * gradient_iterative_2_3

iterative_distance_2_3 = np.linalg.norm(difference_vector_iterative_2_3)
print(f"Iterative Euclidean distance between Image 2 and Image 3: {iterative_distance_2_3}")
```
